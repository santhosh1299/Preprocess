```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("products_data.csv")

```

```python
df.dtypes
```

```python
df.replace(['', ' ', 'NA', 'na', 'N/A', 'n/a', 'NaN', 'nan', '-999', '-9999', 'None', 'none', 'NULL', 'null', 'inf', '-inf', 'Infinity', '-Infinity'], np.nan, inplace=True)
```

```python
df.columns = df.columns.str.strip().str.replace(' ', '_')
```

```python
import pandas as pd

def convert_column_to_dominant_dtype(column):
    # Count occurrences of each data type
    int_count = 0
    float_count = 0
    string_count = 0
    
    for value in column:
        if isinstance(value, int):
            int_count += 1
        elif isinstance(value, float):
            float_count += 1
        elif isinstance(value, str):
            try:
                float(value)  # Check if it can be converted to float
                float_count += 1
            except ValueError:
                string_count += 1
    
    # Determine the dominant data type
    max_count = max(int_count, float_count, string_count)
    if max_count == int_count:
        dominant_dtype = int
    elif max_count == float_count:
        dominant_dtype = float
    else:
        dominant_dtype = str
    
    # Convert the column to the dominant data type
    column = [value if isinstance(value, dominant_dtype) else None for value in column]
    
    return column


# Convert each column to the dominant data type
for column in df.columns:
    df[column] = convert_column_to_dominant_dtype(df[column])

print(df.dtypes)
```
